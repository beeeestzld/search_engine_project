```
import pandas as pd

############################ 加载数据 ############################
df_all = pd.read_csv("drive/My Drive/sougou/data/all_data.csv",encoding="UTF-8",index_col=0)
df_all['Age'] = df_all['Age'] - 1
df_all['Gender'] = df_all['Gender'] - 1
df_all['Education'] = df_all['Education'] - 1
df_all.head()
```

```
df_all.info()
```

```
df_all["Query_num"] = [len(i.split("\t")) for i in df_all["Query_List1"]]
```

```
df_all["Query_num"].describe()
```

```
df_all["Query_len"] = [len(i) for i in df_all["Query_List"]]
df_all["Query_len"] = df_all["Query_len"] - df_all["Query_num"] + 1
```

```
df_all["Query_len"].describe()
```

```
df_all["Ave_query_len"] = df_all["Query_len"]/df_all["Query_num"]
```

```
df_all["Ave_query_len"].describe()
```

```
df_all.head()
```

```
a = ["aa","bbb"]
max([len(b) for b in a])
```

```
query_list = [i.split("\t") for i in df_all["Query_List1"]]
```

```
df_all["Max_query_List1"]=[max([len(k) for k in i]) for i in query_list]
```

```
df_all.head()
```

```
df_all["Space_num"] = [i.count(" ") for i in df_all["Query_List1"]]
```

```
from collections import Counter
df_all["Url_num"] = [Counter(i)["网址"] for i in query_list]
```

```
df_all.head()
```

```
df_all["Man_num"] = [i.count("男") for i in df_all["Query_List1"]]
df_all["Woman_num"] = [i.count("女") for i in df_all["Query_List1"]]
```
